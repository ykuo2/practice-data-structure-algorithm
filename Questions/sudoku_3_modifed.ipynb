## ALL FUNCTIONS

```python
def display_control(p):
    print(display(p))
    clear_output(wait='False')
    time.sleep(0.01)
```

```python
# heuristic 1
def heuristic(plate, location, display_on=None):
    p = [[v for v in x] for x in plate]
    
    # 不去影響之前的數值，不然容易出錯
    for current_location in range(location,81):
        r,c = divmod(current_location,9)
        ava_num = check_available_number(p,r,c)
        if len(ava_num)==1:
            p[r][c]=ava_num[0]
            
            # 顯示解題過程
            if display_on:
                display_control(p)
            # 持續看有沒有其他的直覺解，直到沒有直覺解了為止
            p = heuristic(p, location, display_on)
            return p
    # 如果沒有直覺解了，回傳
    return p
```

```python

from IPython.display import clear_output
import time

def solve(plate, location=0, display_on=None, heuristic_on=None):
    #plate = [[col for col in row] for row in plate]
    # 顯示解題過程
    if display_on:
        display_control(plate)
    
    # 檢查如果解完，則回傳 plate
    if is_complete(plate):
        return plate
    
    # 如果 location>80，則結束之
    if location>80:
        return plate
    
    
    # 如果仍未填入數字的區域無候選清單，則無解
    if is_alive(plate)==False:
        return False
    
    # 將 locaiton 0-80 解析為二維址
    i,j = divmod(location,9)
    
    # 檢查可執行的步驟
    ava_num = check_available_number(plate,i,j) 
    
    # 如果這一步已經有填入數字，則往下一個 location 
    if len(ava_num)==0:
        return_plate = solve(plate, location+1, display_on, heuristic_on)
        if return_plate==False:
            return False
    else:
        # 如果尚未填入數字，從 candidate 中，依序一個一個填入
        for value in ava_num:
            # 在 location 處填入 candidate 數值
            
            #plate = [[col for col in row] for row in plate]
            # 賦值，這改動大幅減少所需時間
            # 這邊可以直接賦值，不需要在最上面或是這邊重新建立plate物件
            # 因為本來就一定要改變數值
            plate[i][j] = value
            # ！！這裡改成 return_plate 不行！！！
            
            # 填入之後，繼續填入下一 location 的數值
            # 如果是無解，則回傳 False　有解回傳 plate
            if heuristic_on:
                # 執行直覺解        
                # 這裡呼叫了 solve，就重新產生一個 matrix，所以不會取代掉本來的 plate
                return_plate = heuristic(plate, location, display_on)
            else:
                # 執行一般 exhausted solution   
                return_plate = [[col for col in row] for row in plate]
            
            # slower
            #return_plate = heuristic_box(return_plate, location)    
            #return_plate = heuristic_rc(return_plate, location)    
            
            # 執行完 heuristic，接著再往下一個解
            return_plate = solve(return_plate, location+1, display_on, heuristic_on)

            # 如果最後回傳為 False -> 則下一個 candidate
            if return_plate==False:
                continue
            else:
            # 如果回傳為 plate -> 代表有解 -> 繼續回傳
                return return_plate
    
    # 若有解，由此回傳
    return return_plate
    
```

```python
def check_available_number(plate, row_location, col_location):
    # 根據傳入的 plate，產生該 location 可行的 ava_num
    # 這個方法比一口氣直接計算所有的可行點然後 iterate 還要慢，但是比較彈性
    # 使用 remove is much more faster than set1-set2 method (about 10x)
    
    # 已經有數字了
    if plate[row_location][col_location]!=0:
        return []
    
    ava_num = [1,2,3,4,5,6,7,8,9]
    # 移除同 row
    for i in range(9):
        try:
            ava_num.remove(plate[row_location][i])
        except:
            pass
    # 移除同 column
    for i in range(9):
        try:
            ava_num.remove(plate[i][col_location])
        except:
            pass
    # 移除同 3x3 box
    row_base = row_location//3*3
    col_base = col_location//3*3
    for r in range(row_base, row_base+3):
        for c in range(col_location, col_base+3):
            try:
                ava_num.remove(plate[r][c])
            except:
                pass
            
    return ava_num
    
```

```python
def new_plate(plate, location, value):
    # 產生一個新的 plate 才不會跟前一個 plate 互相影響到
    x = [[col for col in row] for row in plate]
    
    # 賦值
    i,j = divmod(location,9)
    x[i][j]=value
    return x


def display(plate):
    if type(plate)==bool:
        print('No solution is found')
        return
    # construct each row numbers
    p = []
    for i in range(len(plate)):
        add_row = [' '+str(n)+' ' if n!=0 else '   ' for n in plate[i]]
        p.append(add_row)
        
    # add border
    row_str = []
    for i in range(len(plate)):
        add_str = '|' + '|'.join(p[i]) + '|'
        row_str.append(add_str)
        
    ret =\
    f"""
    -------------------------------------
    {row_str[0]}
    -------------------------------------
    {row_str[1]}
    -------------------------------------
    {row_str[2]}
    -------------------------------------
    {row_str[3]}
    -------------------------------------
    {row_str[4]}
    -------------------------------------
    {row_str[5]}
    -------------------------------------
    {row_str[6]}
    -------------------------------------
    {row_str[7]}
    -------------------------------------
    {row_str[8]}
    -------------------------------------"""
    return ret
    

def is_complete(plate):
    # 判斷是否得解
    n = len(plate)
    
    # rows
    for i in range(n):
        temp_row = set(plate[i])
        # 如果有 0 ，代表尚未完成 
        if 0 in temp_row:
            return False
        if len(temp_row)!=n:
            return False
        
    # cols
    for i in range(n):
        temp_col = set([x[i] for x in plate])
        if 0 in temp_col:
            return False
        if len(temp_col)!=n:
            return False
        
    # 3x3s
    for i in range(0,9,3):
        for j in range(0,9,3):
            temp_9_box = set([plate[x][y] for x in range(i,i+3) for y in range(j,j+3)])
            if 0 in temp_9_box:
                return False
            if len(temp_9_box)!=n:
                return False
    return True

def is_alive(plate):
    for i in range(9):
        for j in range(9):
            # 如果有還沒被填入數值的地方 (value==0)，卻沒有 candidate 了，表示此解不通
            if plate[i][j]==0 and len(check_available_number(plate,i,j))==0:
                return False
                            
    # 如果有重複數值，表示錯誤
    for itm in plate:
        for i in range(1,10):
            if itm.count(i)>1:
                return False
    for col in range(9):
        temp_col = [plate[row][col] for row in range(9)]
        for i in range(1,10):
            if temp_col.count(i)>1:
                return False
    # 3x3 box
    for r in range(3):
        for c in range(3):
            temp_box = [plate[3*r+i][3*c+j] for i in range(3) for j in range(3)]
            for i in range(1,10):
                if temp_box.count(i)>1:
                    return False
                    
    return True


    
```
